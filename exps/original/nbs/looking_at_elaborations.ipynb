```python
df = pd.read_csv('/Users/kate_fieseler/PycharmProjects/syndirella/exps/original/KFKUHRDIQZWKPK-UHFFFAOYSA-N_Sp2-sp2_Suzuki_coupling_products_1of1_no_non_mcs_match_placements.csv', index_col=0)
df
```

```python
# how many acceptable?
df = df[~df['outcome'].isna()]
df
```

```python
df['outcome'].value_counts()
```

```python
good = df[df['good_expansion'] == True]
good
```

```python
good['outcome'].value_counts() / len(good)
```

```python
bad = df[df['good_expansion'] == False]
bad
```

```python
# how many acceptable?
bad['outcome'].value_counts() / len(good)
```

```python
import matplotlib.pyplot as plt

good_data = good['∆∆G']
bad_data = bad['∆∆G']

plt.hist(good_data, bins=20, edgecolor='green', linewidth=1.5, facecolor='none', label='Good')
plt.hist(bad_data, bins=20, edgecolor='red', linewidth=1.5, facecolor='none', label='Bad')

plt.xlabel('∆∆G')
plt.ylabel('Frequency')
plt.legend()

plt.show()
```

```python
# Assuming 'good' and 'bad' are your dataframes and 'num_atom_diff' is the column of interest
good_data = good['comRMSD']
bad_data = bad['comRMSD']

plt.hist(good_data, bins=20, edgecolor='green', linewidth=1.5, facecolor='none', label='Good')
plt.hist(bad_data, bins=20, edgecolor='red', linewidth=1.5, facecolor='none', label='Bad')

plt.xlabel('RMSD')
plt.ylabel('Frequency')
plt.legend()

plt.show()
```
