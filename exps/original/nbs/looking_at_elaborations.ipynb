```python
import pandas as pd
```

```python
df = pd.read_csv('/Users/kate_fieseler/PycharmProjects/syndirella/exps/original/KFKUHRDIQZWKPK-UHFFFAOYSA-N_Sp2-sp2_Suzuki_coupling_products_1of1_no_non_mcs_match_placements.csv', index_col=0)
df
```

```python
# how many placed?
df = df[~df['outcome'].isna()]
df
```

```python
# make corrected delta delta G column
df['∆∆G_rot_bond'] = df['∆∆G'] + df['N_rotatable_bonds']
```

```python
# add outcome column including new delta delta G
df['outcome_new'] = df['outcome']
df.loc[(df['∆∆G_rot_bond'] < 0) & (df['comRMSD'] < 2), 'outcome_new'] = 'acceptable'
df.loc[(df['∆∆G_rot_bond'] < 0) & (df['∆∆G_rot_bond'] > -1) & (df['outcome'] == 'acceptable'), 'outcome_new'] = 'weak'
df.loc[(df['∆∆G_rot_bond'] > 0), 'outcome_new'] = 'too contorted'
df
```

```python
good = df[df['good_expansion'] == True]
good
```

```python
bad = df[df['good_expansion'] == False]
bad
```

```python
# only remove NaN '∆∆G_rot_bond' values
good = good[~good['∆∆G_rot_bond'].isna()]
good
```

```python
bad = bad[~bad['∆∆G_rot_bond'].isna()]
bad
```

```python
import matplotlib.pyplot as plt
import numpy as np

good_data = good['∆∆G_rot_bond']
bad_data = bad['∆∆G_rot_bond']

plt.hist(good_data, histtype='step', bins=20, edgecolor='#228833', linewidth=1.5, facecolor='none', label='Good')
plt.hist(bad_data, histtype='step', bins=20, edgecolor='#EE6677', linewidth=1.5, facecolor='none', label='Bad')

# Calculate and plot medians
median_good = np.median(good_data)
print(median_good)
median_bad = np.median(bad_data)
print(median_bad)

plt.axvline(median_good, color='#228833', linestyle='dashed', linewidth=1.5, label='Median Good')
plt.axvline(median_bad, color='#EE6677', linestyle='dashed', linewidth=1.5, label='Median Bad')

plt.xlabel('∆∆G with rotatable bond penalty')
plt.ylabel('Frequency')
plt.legend()

plt.show()
```
