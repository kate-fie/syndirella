```python
# Make master csv of all base compounds
columns = ['smiles', 'submitter_name', 'ref_pdb', 'hit_names', 'ID', 'ROMol']
def convert_csv_to_columns(df):
    if 'smiles' in df.columns:
        df.drop(columns=['smiles'], inplace=True)
    df = df.rename(columns={'original SMILES': 'smiles', 'submitter_name': 'submitter_name', 'ref_pdb': 'ref_pdb', 'ref_mols': 'hit_names', 'ID': 'ID', 'ROMol':'ROMol'})
    if 'cmpd_id' in df.columns:
        columns.append('cmpd_id')
    # remove columns not in columns list
    df = df[columns]
    # fill rows with first value in submitter_name column
    df['submitter_name'] = df['submitter_name'].fillna(method='ffill')
    # remove duplicate columns
    df = df.loc[:,~df.columns.duplicated()].copy()
    #df = df.drop_duplicates(subset=['smiles'])
    # remove any row with smiles NaN
    df = df[df['smiles'].notna()]
    return df
```

```python
# check if any smiles in smiles column are duplicated by tanimoto similarity of 1
import pandas as pd
from rdkit import Chem
from rdkit.Chem import AllChem
from rdkit.DataStructs import BulkTanimotoSimilarity

def find_duplicate_smiles(df, threshold=0.85):
    # Convert SMILES to molecular fingerprints
    fingerprints = [AllChem.GetMorganFingerprint(Chem.MolFromSmiles(smile), 2) for smile in df['smiles'] if smile]

    # Initialize a list to keep track of duplicates
    duplicate_indices = set()

    # Calculate Tanimoto similarity and identify duplicates
    for i in range(len(fingerprints)):
        for j in range(i + 1, len(fingerprints)):
            similarity = BulkTanimotoSimilarity(fingerprints[i], [fingerprints[j]])[0]
            if similarity >= threshold:
                duplicate_indices.add(i)
                duplicate_indices.add(j)

    # Create a DataFrame from the duplicate indices
    duplicates_df = df.iloc[list(duplicate_indices)].drop_duplicates()

    return duplicates_df
```

```python
# Example usage
duplicates = find_duplicate_smiles(master_df, threshold=1)
duplicates.to_csv('base_compound_duplicates.csv', index=False)
```
