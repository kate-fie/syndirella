```python

import json
```

```python
with open("../data/dev/grouped_smirks_with_codes.json", 'r') as f:
    grouped_smirks = json.load(f)
len(grouped_smirks)
```

```python
with open('../data/2025-04-04_rxn_classifer_91_test_data_labeled.json', 'r') as f:
    labels = json.load(f)
len(labels)
```

```python
codes: dict = {}
for rxn in labels:
    try:
        print(rxn['label'])
        print(rxn['route_metadata']['template_code'])
        code = rxn['route_metadata']['template_code']
        codes[code] = {'label': rxn['label'], 'predicted': [], 'sim': []}
        # search through grouped_smirks to see what reactions this label corresponds with, calculate accuracy
        for group in grouped_smirks:
            if not group['uspto_templates']:
                continue
            for template in group['uspto_templates']:
                if template['template_code'] == code:
                    codes[code]['predicted'].append(group['name'])
                    codes[code]['sim'].append(template['rxn_sim'])
            for match in group['matched']:
                if not match['uspto_templates']:
                    continue
                for template in match['uspto_templates']:
                    if template['template_code'] == code:
                        codes[code]['predicted'].append(match['name'])
                        codes[code]['sim'].append(template['rxn_sim'])
    except KeyError:
        print(rxn)
```

```python
codes
```

```python
# Compute accuracy of predicted labels containing the correct label
correct_predictions = 0
total_labels = len(codes)

for code, data in codes.items():
    if (data['label'] in data['predicted']) or (data['label'] == '' and len(data['predicted']) == 0):
        correct_predictions += 1

accuracy = correct_predictions / total_labels if total_labels > 0 else 0
print(f"Accuracy: {accuracy:.2%}")
```

```python
# redo with counting overarching Syndirella naming in predicted if sub is included
codes: dict = {}
for rxn in labels:
    try:
        print(rxn['label'])
        print(rxn['route_metadata']['template_code'])
        code = rxn['route_metadata']['template_code']
        codes[code] = {'label': rxn['label'], 'predicted': [], 'sim': []}
        # search through grouped_smirks to see what reactions this label corresponds with, calculate accuracy
        for group in grouped_smirks:
            if not group['uspto_templates']:
                continue
            for template in group['uspto_templates']:
                if template['template_code'] == code:
                    codes[code]['predicted'].append(group['name'])
                    codes[code]['sim'].append(template['rxn_sim'])
            for match in group['matched']:
                if not match['uspto_templates']:
                    continue
                for template in match['uspto_templates']:
                    if template['template_code'] == code:
                        codes[code]['predicted'].append(match['name'])
                        codes[code]['sim'].append(template['rxn_sim'])
                        if group['name'] not in codes[code]['predicted']:
                            codes[code]['predicted'].append(group['name'])
                            codes[code]['sim'].append(template['rxn_sim'])
    except KeyError:
        print(rxn)
```

```python
correct_predictions = 0
total_labels = len(codes)
wrong_predictions = []

for code, data in codes.items():
    if (data['label'] in data['predicted']) or (data['label'] == '' and len(data['predicted']) == 0):
        correct_predictions += 1
    else:
        wrong_predictions.append({code: data})

accuracy = correct_predictions / total_labels if total_labels > 0 else 0
print(f"Accuracy: {accuracy:.2%}")
```
