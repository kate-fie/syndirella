```python
import json
import pandas as pd

# Load the best classifier results
with open('../data/dev/grouped_smirks_with_codes_concat_True_fp_maccs_threshold_0.2_results.json', 'r') as f:
    classifier_data = json.load(f)

# Filter for the best classifier (method = 'all', threshold = 0.2)
best_classifier_data = [
    entry for entry in classifier_data if entry.get('method') == 'all' and entry.get('threshold') == 0.2
]
per_rxn_df = pd.DataFrame(best_classifier_data[0]['per_reaction_results'])

# Load the test data
with open('../data/2025-04-04_rxn_classifer_91_test_data_labeled.json', 'r') as f:
    test_data = json.load(f)
test_df = pd.DataFrame(test_data)

# Reset the index of test_df and rename it to match the key in per_rxn_df
test_df = test_df.reset_index().rename(columns={'index': 'rxn_i'})

# Merge the two DataFrames on 'rxn_i'
df = pd.merge(per_rxn_df, test_df, on='rxn_i', how='inner')

# Display the merged DataFrame
df
```
