```python
# show distributions of num atom diff of good and bad expansions --> to show we have a even distribution of both
import matplotlib.pyplot as plt

# Assuming 'good' and 'bad' are your dataframes and 'num_atom_diff' is the column of interest
good_data = good['num_atom_diff']
bad_data = bad['num_atom_diff']

plt.hist(good_data, bins=20, edgecolor='green', linewidth=1.5, facecolor='none', label='Good')
plt.hist(bad_data, bins=20, edgecolor='red', linewidth=1.5, facecolor='none', label='Bad')

plt.xlabel('num_atom_diff')
plt.ylabel('Frequency')
plt.legend()

plt.show()
```
