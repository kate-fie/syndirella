```python
def compare_lists_of_dicts(list1, list2):
    differences = []

    # Ensure we have lists of equal length
    for i in range(max(len(list1), len(list2))):
        dict1 = list1[i] if i < len(list1) else 'No entry in list1'
        dict2 = list2[i] if i < len(list2) else 'No entry in list2'
        
        # If one of the lists is shorter, record the missing entry
        if isinstance(dict1, str) or isinstance(dict2, str):
            differences.append((i, dict1, dict2))
            continue

        # If both are dictionaries, compare key by key
        for key in set(dict1.keys()) | set(dict2.keys()):
            val1 = dict1.get(key, 'Key not in dict1')
            val2 = dict2.get(key, 'Key not in dict2')
            if val1 != val2:
                differences.append((i, key, val1, val2))

    return differences

# Example usage:
list1 = ast.literal_eval(df['reactant1_metadata'][5])

list2 = ast.literal_eval(df['reactant1_metadata'][6])

diffs = compare_lists_of_dicts(list1, list2)

for diff in diffs:
    if len(diff) == 3:
        print(f"Index {diff[0]}: Missing entry - {diff[1]}")
    else:
        print(f"Index {diff[0]} Key {diff[1]}: {diff[2]} vs {diff[3]}")
```

```python
num_analogs_placed = sum([min(x,10000) for x in df.loc[df['place_done'] == True, 'Analogs']])
num_analogs_placed
```

```python
# How many total placements are there?
num_analogs = sum(min(x, 10000) if pd.notna(x) else 0 for x in df['Analogs'])
num_analogs
```

```python
# What percentage of placements have happened for 1-step reactions?
num_analogs_placed/num_analogs
```

```python
# How many placements are successful?
total_success = sum([x for x in df['success_place']])
total_success
```

```python
# What percentage of placements are successful?
total_success/num_analogs_placed
```

```python
# Per compound, what is the average ratio of successful placements to total placements if they were placed?
# Get df with only compounds that were placed
df_placed = df.loc[df['place_done'] == True]
df_placed
```

```python
# Per compound, what is the distribution of the ratio of successful placements to total placements if they were placed? (only 10K per elaboration)
# Make new column with ratio of successful placements to total placements (only 10K per elaboration)
df_placed['number_placed'] = df_placed['Analogs'].apply(lambda x: min(x, 10000))
df_placed['ratio_success'] = df_placed['success_place']/df_placed['number_placed']
```

```python
df_placed
```

```python
# Distribution of ratio of successful placements to total placements (only 10K per elaboration)
ax = df_placed.ratio_success.hist()
ax.set_xlabel('Ratio of successful placements to total placements per base compound')
# Annotate each bin
for rect in ax.patches:
    height = rect.get_height()
    ax.annotate(f'{int(height)}', (rect.get_x() + rect.get_width() / 2, height), 
                ha='center', va='bottom')
```

```python
# What is the average ratio of successful placements to total placements if they were placed?
df_placed['ratio_success'].mean()
```

```python
# What is the count of analogs per reaction?
df_with_analogs = df.loc[df['Analogs'].notna() & (df['Analogs'] > 0)]
df_with_analogs
```
