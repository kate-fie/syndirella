```python
# Distribution of RMSD 
ax = df.comRMSD.hist()
ax.set_xlabel('RMSD')
# Annotate each bin
for rect in ax.patches:
    height = rect.get_height()
    ax.annotate(f'{int(height)}', (rect.get_x() + rect.get_width() / 2, height), 
                ha='center', va='bottom')
```
