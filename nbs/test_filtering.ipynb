```python
from rdkit.Chem.FilterCatalog import *
import pandas as pd
from rdkit import Chem
import ast
```

```python
df = pd.read_csv('/Users/kate_fieseler/PycharmProjects/syndirella/exps/original/extra/KFKUHRDIQZWKPK-UHFFFAOYSA-N_Sp2-sp2_Suzuki_coupling_r1_1of1.csv', index_col=0)
df
```

```python
# Initialize FilterCatalogParams
params = FilterCatalogParams()
params.AddCatalog(FilterCatalogParams.FilterCatalogs.PAINS_A) # BRENK is same as Dundee I think
# Create the FilterCatalog
catalog = FilterCatalog(params)
for i in range(catalog.GetNumEntries()):
    entry = catalog.GetEntry(i)
    print(f"Entry {i}:")
    print("  Description:", entry.GetDescription())
```

```python
molecules = [Chem.MolFromSmiles(smi) for smi in df['r1_smiles']]
molecules
```

```python
# Filter molecules
filtered_molecules = []
flagged_molecules = []
for mol in molecules:
    if not catalog.HasMatch(mol):
        filtered_molecules.append(mol)
    else:
        flagged_molecules.append(mol)
```
