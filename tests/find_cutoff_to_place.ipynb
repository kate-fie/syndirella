```python
# make histogram of the number of acceptable compounds and num atom difference
accept_df = df_full[df_full['outcome'] == 'acceptable']
accept_df
```

```python
ax = accept_df['num_atom_difference'].hist(bins=10)
# Annotate each bin
for rect in ax.patches:
    height = rect.get_height()
    ax.annotate(f'{int(height)}', (rect.get_x() + rect.get_width() / 2, height), 
                ha='center', va='bottom')
```
