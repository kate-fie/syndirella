```python
df = pd.read_csv('/Users/kate_fieseler/PycharmProjects/retrievesynthesizable/A71EV2A/routes_data/DEC5/DEC5_routes_syndirella_1step_OUTPUT.csv', index_col=0)
df
```

```python
# Enumerate through the parent directory
parent_directory = '/Users/kate_fieseler/PycharmProjects/retrievesynthesizable/A71EV2A/elabs/1_step'  # Update this path
suffix = '1_of_1.csv'
# Replace all colons with underscores
df['dir_name'] = df['dir_name'].str.replace(':', '_')
for root, dirs, files in os.walk(parent_directory):
    for directory in dirs:
        for index, row in df.iterrows():
            dir_names = [row['dir_name']]
            print(index)
            print(dir_names)
            if directory in dir_names:
                # If directory exists, go into the directory and find .csv
                csv_files = [f for f in os.listdir(os.path.join(root, directory)) if f.endswith(suffix)]
                for csv_file in csv_files:
                    try:
                        analog_number = int(csv_file.split('.')[0].split('_')[-4])  # Extract the integer
                        df.at[index, 'Analogs'] = analog_number
                        df.at[index, 'Done_Time'] = datetime.datetime.now().strftime('%Y-%m-%d %H:%M:%S')
                    except ValueError:
                        pass
                break
```
